```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def splitter(tweet):
    return tweet.split()

vectorizer = CountVectorizer(analyzer=splitter).fit(df["cleaned_tweet"])
vocabs = vectorizer.vocabulary_
```
