# Data Preparing

```python
df_gabung = pd.concat([df_gunung, df_pantai, df_kuliner], ignore_index=True)
```

```python
df_gabung
```

# TF-IDF part

```python
def splitter(tweet):
    return tweet.split()
```

```python
vectorizer = CountVectorizer(analyzer=splitter).fit(df_gabung["cleaned_tweet"])
```

```python
xTarget = vectorizer.transform(df_gabung["cleaned_tweet"])

print(len(vectorizer.vocabulary_))
print(xTarget.shape)
```

```python
tfidf = TfidfTransformer()
X = tfidf.fit_transform(xTarget)

print(X.shape)
print(X)
```

# Visualization

```python
#create multidimensional array
tfidf_array = X.toarray()
tfidf_lists = tfidf_array[:3000]
```

```python
#remove zeros from array
tfidf_lists_nozero = []

for array in tfidf_lists:
    temp_array = []
    for data in array:
        if data != 0:
            temp_array.append(data)
    tfidf_lists_nozero.append(temp_array)
```

```python
tfidf_lists_nozero
```

```python
#takes 5 first words
tfidf_lists_first5 = []

for array in tfidf_lists_nozero:
    tfidf_lists_first5.append(array[:5])
```

```python
tfidf_lists_first5
```

```python
columns = ["tfidf_kata1", "tfidf_kata2", "tfidf_kata3", "tfidf_kata4", "tfidf_kata5"]
df_tfidf = pd.DataFrame(tfidf_lists_first5, columns=columns)
```

```python
df_tfidf.tail()
```

```python
df_visual["tweet"] = df_gabung["cleaned_tweet"]
df_visual["tfidf kata 1"] = df_tfidf["tfidf_kata1"]
df_visual["tfidf kata 2"] = df_tfidf["tfidf_kata2"]
df_visual["tfidf kata 3"] = df_tfidf["tfidf_kata3"]
df_visual["tfidf kata 4"] = df_tfidf["tfidf_kata4"]
df_visual["tfidf kata 5"] = df_tfidf["tfidf_kata5"]
df_visual["label"] = df_gabung["label"]
```
